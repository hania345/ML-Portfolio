```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming df is loaded correctly from your CSV file
df = pd.read_csv("C:/Users/aksha_t30bq4h/Desktop/New folder/ml/Data_Train.csv")

# Check the data types of columns
print(df.dtypes)

# Exclude non-numeric columns from correlation computation
numeric_columns = df.select_dtypes(include='number').columns
correlation_df = df[numeric_columns].corr()

# Plotting heatmap of correlations
plt.figure(figsize=(10, 6))
sns.heatmap(correlation_df, annot=True, fmt='.2f', cmap='Blues')
plt.title('Correlation Heatmap')
plt.show()
```

```python
# Convert duration to a more usable format (minutes)
def duration_to_minutes(duration):
    h, m = 0, 0
    if 'h' in duration:
        h = int(duration.split('h')[0].strip())
    if 'm' in duration:
        m = int(duration.split('m')[0].split()[-1].strip())
    return h * 60 + m

df['Duration_Minutes'] = df['Duration'].apply(duration_to_minutes)

# Set the size of the plot
plt.figure(figsize=(14, 7))

# Create a histogram for flight durations
sns.histplot(df['Duration_Minutes'], kde=True, bins=50)
plt.title('Flight Duration Analysis')
plt.xlabel('Duration (Minutes)')
plt.ylabel('Frequency')

# Display the plot
plt.show()
```
